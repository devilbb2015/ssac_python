```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
import requests # 서버를 위한 requests
```

```python
source = requests.get('https://www.weather.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(source.content, "html.parser")
```

```python
table = soup.find('table',{'class':'table_develop3'})
data = []
```

```python
#[Code 0]
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            location = td.find('a').text # 위치
            temp = tds[5].text # 기온
            humidity = tds[9].text # 습도
            # print(location, temp, humidity)
            print("{0:<7} {1:<7} {2:<7}".format(location, temp, humidity))
            # {0:<10} 이 뜻하는 것은 {0} 값을 ":<10" 10자리로 표현할건데 왼쪽 정렬을 할 것 이다. 라는 뜻
            # {1:>5} 가 뜻하는 것은 {1} 값을 ":>5" 5자리로 표현할건데 오른쪽 정렬을 할 것 이다. 라는 뜻
            
            data.append([location, temp, humidity])

#출력 종료를 표시하기!
print("#"*30)
print("\n감사합니다. 좋은 하루 보내세요!\n")
print("#"*30)

print(data)

with open('weather.csv', 'w') as f:
    f.write('지역, 온도, 습도\n')
    for i in data:
        f.write('{0},{1},{2}\n'.format(i[0],i[1],i[2]))
        
df = pd.read_csv('weather.csv', index_col='지역', encoding='euc-kr')

city_df = df.loc[['서울', '인천', '대전', '대구', '광주', '부산', '울산']]
```
