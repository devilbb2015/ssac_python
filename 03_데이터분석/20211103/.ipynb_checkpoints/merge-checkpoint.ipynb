```python
import pandas as pd
import numpy as np
```

```python
cust = pd.DataFrame({
    '고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름' : ['가', '나', '다', '라', '마', '바', '사']
}, columns=['고객번호', '이름'])
cust
```

```python
order = pd.DataFrame({
    '고객번호' : [1001, 1001, 1005, 1007, 1008, 1001],
    '금액' : [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
order
```

```python
pd.merge(cust, order)
```

```python
pd.merge(cust, order, how='outer')
```

```python
pd.merge(cust, order, how = 'left')
```

```python
pd.merge(cust, order, how = 'right')
```

```python
df1 = pd.DataFrame({
    '품종' : ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이' : [1.4, 1.3, 1.5, 1.3]
}, columns = ['품종', '꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종' : ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비' : [0.4, 0.3, 0.5, 0.3]
}, columns = ['품종', '꽃잎너비'])
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
df1 = pd.DataFrame({
    '고객명' : ['춘향', '춘향', '몽룡'],
    '날짜' : ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터' : ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명' : ['춘향', '몽룡'],
    '데이터' : ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2, on = "고객명")
```

```python
df1 = pd.DataFrame({
    '이름' : ['영희', '철수', '철수'],
    '성적' : [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명' : ['영희', '영희', '철수'],
    '성적2' : [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on ='이름', right_on = '성명')
```
