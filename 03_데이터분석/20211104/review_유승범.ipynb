```python
import pandas as pd
import numpy as np
```

```python
url = 'CARD_SUBWAY_MONTH_202102.csv'

data = pd.read_csv(url)
df = pd.DataFrame(data)

df
```

```python
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year+'-'+month+'-'+day
```

### df[컬럼명].apply(함수이름)

```python
df['사용일자1'] = df['사용일자'].apply(getDate)
df.head()
```

### merge

```python
df1 = pd.DataFrame({'이름': ['A', 'B', 'C'],
                   '성별': ['남자', '여자', '남자']})

df1
```

```python
df2 = pd.DataFrame({'이름': ['A', 'B', 'D'],
                   '키': [180, 160, 150]})

df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

### pd.concat([데이터프레임], axis= )

```python
df1 = pd.DataFrame({'이름': ['A', 'B', 'C', 'D', 'E'],
                   '키': [180, 160, 150, 170, 170],
                   '나이': [23, 46, 33, 65, 28],
                   '주소': ['서울', '부산', '대전', '원주', '제주'],})

df1
```

```python
df2 = pd.DataFrame({'이름': ['F', 'G', 'H', 'I', 'J'],
                   '키': [190, 180, 150, 160, 170],
                   '나이': [22, 18, 87, 45, 74],
                   '주소': ['전주', '서울', '수원', '부산', '인천'],})

df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.concat([df1, df2], axis=0)
```
