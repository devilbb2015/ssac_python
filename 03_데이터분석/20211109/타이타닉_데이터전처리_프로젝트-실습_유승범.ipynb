## 데이터 설명 : 
- Survived : 생존 여부
        0 = 사망
        1 = 생존
- Pclass : 객실 등급
        1st = Upper
        2nd = Middle
        3rd = Lower
- Name : 이름
- Sex : 성별
- Age : 나이
- Siblings/Spouses Aboard : 동반한 Sibling(형제자매)와 Spouse(배우자)의 수
- Parents/Children Aboard : 동반한 Parent(부모) Child(자식)의 수
- Fare : 티켓의 요금 (USD)

# 연령대 컬럼을 추가하시오 
컬럼이름: AgeGroup

연령대 : 미성년자, 20대, 30대, 40대, 50대, 60대이상 : 'Minor', '20s', '30s', '40s', '50s', '60+'

```python
df['AgeGroup'] = pd.cut(df.Age, bins=[0, 20, 30, 40, 50, 60, 100], labels=['Minor', '20s', '30s', '40s', '50s', '60+'])
```

```python
df
```

# 각 연령대별 인구 수를 구하시오 

```python
df.AgeGroup.value_counts().sort_index()
```

# 각 승객의 Fare를 FareinWon이라는 원화 컬럼을 추가하시오 
- 1 USD = 1150 KRW 

```python
df['FareinWon'] = df['Fare'] * 1150
```

# 'Gender'라는 컬럼을 추가하여 'Sex'컬럼의 male과 female을 0과 1로 변환하시오 

```python
df['Gender'] = df['Sex'].map({'female':1, 'male':0}).astype(int)
df
```

# 승객들의 이름에서 타이틀(Mr, Mrs, Miss, Ms., Rev. 등 )를 제거하시오.

```python
df.Name.str.replace('^(\w+\.\W)','')
```

```python
df['Name_NoTitle'] = df.Name.str.replace('^(\w+\.\W)','')
df.head()
```

```python
df['Name_NoTitle'] = df['Name'].apply(lambda x: x.split('.')[1].strip() if '.' in x else x.strip())
df
```

# 데이터셋 안에서 세가지의 분석 포인트를 찾고 결론을 쓰시오. 

```python
df.corr()
```

```python
male_count = df[df['Gender'] == 0]
print("전체 남성 수 :", len(male_count))
female_count = df[df['Gender'] == 1]
print("전체 여성 수 :", len(female_count))
df_male = df[(df['Gender'] == 0) & (df['Survived'] == 1)]
print("생존한 남성 수 :", len(df_male))
df_female = df[(df['Gender'] == 1) & (df['Survived'] == 1)]
print("생존한 여성 수 :", len(df_female))
```

## 남성보다 여성의 생존비율이 높다

```python
df_class3 = df[(df.Pclass == 3) & (df.Survived == 1)]
df_class2 = df[(df.Pclass == 2) & (df.Survived == 1)]
df_class1 = df[(df.Pclass == 1) & (df.Survived == 1)]
print('전체 Lower등급 승객 수', len(df[df.Pclass == 3]),'명중 생존자',len(df_class3),'명')
print('전체 Middle등급 승객 수', len(df[df.Pclass == 2]),'명중 생존자',len(df_class2),'명')
print('전체 Upper등급 승객 수', len(df[df.Pclass == 1]),'명중 생존자',len(df_class1),'명')
```

## 객실등급이 높을수록 생존비율이 높다.

```python
print(len(df))
df_solo = df[(df['Siblings/Spouses Aboard'] == 0) & (df['Parents/Children Aboard'] == 0)]
print('전체 승객 수는',len(df),'혼자 온 승객 수는',len(df_solo))
```
