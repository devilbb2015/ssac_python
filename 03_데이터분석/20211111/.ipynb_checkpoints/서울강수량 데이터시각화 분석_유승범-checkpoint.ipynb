```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df = pd.read_csv('./data/rn_20211111133010.csv', encoding='cp949', skiprows=7)
df.head()
```

```python
df.info()
```

```python
birth = df[df['날짜'].str.contains('-03-25')]
birth.head()
```

```python
df['강수량(mm)'] = df['강수량(mm)'].fillna(0)
```

```python
df
```

```python
df['년도'] = df['날짜'].apply(lambda x : x.split("-")[0].strip() )
df['월'] = df['날짜'].apply(lambda x : x.split("-")[1].strip() )
df
```

```python
df.info()
```

```python
df1 = birth.copy()
df1.head()
```

```python
df1.fillna(0, inplace=True)
df1.head()
```

```python
df1['년도'] = df1['날짜'].apply(lambda x : x.split("-")[0].strip() )
df1['월'] = df1['날짜'].apply(lambda x : x.split("-")[1].strip() )
df1.head()
```

```python
df1.set_index('년도', inplace=True)
df1.head()
```

```python
df1['강수량(mm)'].plot(figsize=(10,4))
```

#### 월별 그룹화

```python
df2 = df.copy()
df2.head()
```

```python
df1['월'] = df1['날짜'].str.slice(5,7).astype(int)
df2['그룹'] = df2['년도']+df2['월']
df2.head()
```

#### 월별 강수량 계산

```python
df2.boxplot(column='강수량(mm)', by='그룹', figsize=(13, 5))
```

```python
df2_sum = df2.groupby('그룹').sum()
df2_sum
```

```python
df2_sum['강수량(mm)'].plot(figsize=(20,6))
```
