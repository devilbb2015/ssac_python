```python
# 워드클라우드 생성을 위한 import 셋팅
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 워드 클라우드 생성을 위한 텍스트 형성
text = '''
python python python python python python python python python python python python python 
big data big data big data big data big data big data big data big data 
bigData bigData bigData bigData bigData bigData bigData bigData bigData bigData bigData 
Crawling Crawling Crawling Crawling Crawling Crawling 
machine machine machine machine machine machine machine machine machine machine 
learning learning learning learning learning 
deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep deep 
강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 강아지 
고양이 고양이 고양이 고양이 고양이 고양이 
기린 기린 기린 기린 기린 기린 기린 기린 기린 기린 기린 
코끼리 코끼리 코끼리 코끼리 
'''

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0, font_path=r'c:\Windows\Fonts\malgun.ttf')
plt.imshow(wc.generate(text))
plt.axis("off")
plt.show()
```
