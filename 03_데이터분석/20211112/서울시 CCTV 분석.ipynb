## CCTV와 인구현황 그래프로 분석하기

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure()

data_result['CCTV'].plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
data_result['CCTV'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

- CCTV 개수는 서울시에 타 지역구보다 강남구가 월등이 많이 설치 되어 있음
- 상대적으로 양천구, 서초구, 은평구도 많은 CCTV 설치가 되어 있음
- 인구대비 CCTV 비율은 추가 분석이 필요함

```python
data_result['CCTV비율'] = data_result['CCTV'] / data_result['인구수'] * 100

data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

- 인구대비 CCTV 수를 분석결과 용산구와 종로구가 월등히 높음
- 송파구는 인구대비 CCTV 비율이 낮음

```python
# - Scatter 함수, s=50으로 마커 크기 설정
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['CCTV'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

## numpy의 polyfit을 활용하여 시각화 분석

```python
fp1 = np.polyfit(data_result['인구수'], data_result['CCTV'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_result['인구수'], data_result['CCTV'], s=50)
# ls = line Style, lw = line Width
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='b')

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

- CCTV와 인구수는 양의 상관관계가 있음

## 조금 더 설득력 있는 자료 만들어보기

```python
# 하나의 직선이 이 전체 데이터의 대표 값 역활을 한다면, 즉 구의 수가 300000 일때는 CCTV는 1100정도
# 중심의 경향에서 멀어져 있는 구를 다른 색으로 표현하기 위해서 오차를 계산하여 오차가 큰 순으로 데이터를 정렬

fp1 = np.polyfit(data_result['인구수'], data_result['CCTV'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100) # x선의 범위, x점의 개수를 선언해줌

data_result['오차'] = np.abs(data_result['CCTV'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
df_sort
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['CCTV'], 
            c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

# 정렬한 데이터 프레임의 plt.text를 활용

for n in range(len(df_sort)):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['CCTV'][n]*0.98, 
             df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```

- 서울시에서 다른 구와 비교했을 때, 강남구, 양천구, 서초구, 은평구는 CCTV가 많지만,
- 송파구, 강서구, 도봉구, 마포구는 다른 구에 비해 CCTV 비율이 낮다
