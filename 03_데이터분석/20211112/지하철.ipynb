```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins
```

```python
# 데이터 불러오기
subway = pd.read_csv('./data/지하철_위경도_처리.csv', encoding='cp949')
subway
```

```python
# 1호선 추출
line1 = subway[subway.선명 == '1호선']
line1
```

```python
# 2호선 추출
line2 = subway[subway.선명 == '2호선']
line2.reset_index(drop=True, inplace=True)
line2
```

```python
# 폴리움

# 지도 기본 위치 세팅
map = folium.Map( location=[37.52389, 126.96585], tiles='OpenStreetMap', zoom_start=11)

# 각 지하철역을 마커를 사용하여 지도에 표시하기
for name, lat, long in zip(line2.역이름, line2.위도, line2.경도):
    folium.Marker([lat, long],
                 tooltip=name,
                 icon=folium.Icon(prefix='fa', icon='subway', color='green')
                 ).add_to(map)
    
# 지도 전체화면
plugins.Fullscreen(position='topright',
                  title='Click to Expand',
                  title_canncel='Click to Exit',
                  force_separate_button=True).add_to(map)

# 지도스크롤
plugins.MousePosition().add_to(map)

# 지도 띄우기
map
```
