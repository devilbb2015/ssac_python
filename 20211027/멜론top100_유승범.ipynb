```python
# 크롤링을 위한 기본 셋팅 import

from bs4 import BeautifulSoup
from selenium import webdriver # webdriver를 위해서 selenium 필요

import pandas as pd
from time import sleep
```

```python
# 웹 드라이버 실행
driver = webdriver.Chrome("chromedriver")
```

```python
# 사이트 접속한다.
sleep(0.5)
driver.get('https://www.melon.com/chart/index.htm')
```

```python
# bs4 쓰려면 페이지소스를 가져와서 작업하기
# driver.page_source
# make bs4 object
soup = BeautifulSoup(driver.page_source)
soup
```

```python
# 노래제목 가져오기
# 멜론의 현재(2021.10.27 10:00) 순위 확인
# 홈페이지의 태그 정보 확인
# Top 100 노래를 Crawling 하기

rank_title = []
rank_musician = []
#rank = range(1, 101)
rank_rank = []

title = soup.select('div.rank01 > span > a')
musician = soup.select('div.rank02 > a')
rank = soup.select('span.rank')

for i in title:
    rank_title.append(i.text)
       
for j in musician:
    rank_musician.append(j.text)

for k in rank[1:]:
    rank_rank.append(k.text)
    
rank_li = list(zip(rank_rank, rank_title, rank_musician))

for li in rank_li:
    print(li)
```
