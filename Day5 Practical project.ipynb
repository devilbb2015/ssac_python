## 2. number1.txt 파일을 불러와서 파일 안의 coffee와 tea 단어 개수를 구하시오 

```python
f2 = open("number1.txt", 'r')
# 파일의 내용을 읽어오고 싶어 = readline()
line = f2.readline().split(' ')
print("coffee", line.count('coffee'))
print("tea", line.count('tea'))
```

```python
dic = {}

for i in list(set(line)):
    dic[i] = 0
    
for item in line:
    if item == list(set(line))[0]:
        dic[list(set(line))[0]] +=1
    if item == list(set(line))[1]:
        dic[list(set(line))[1]] +=1
            
print(dic)
```

## 유리병 안에 7개의 빨간색 구슬과 12개의 파랑색 구슬이 들어있다. 사용자의 인풋을 받아 유리병에서 꺼낼 구슬의 색깔을 정하고, 병 안에 남은 구슬의 개수를 프린트하는 함수를 만드시오.
(사용자는 총 다섯개의 구슬을 꺼낼 수 있으며 인풋이 빨강/파랑이 아닐 경우 에러메세시를 출력하시오.) 


```python
def beads():
    red=7
    blue=12
    for i in range(5):
        color = input("색 :")
        while True:
            if color == "빨강":
                red -= 1
            elif color == "파랑":
                blue -=1
            else:
                print("입력값이 잘못되었습니다.")
            print("빨강색 구슬은", red ,", 파랑색 구슬은", blue)
            break
        
        
```

```python
beads()
```

```python
def factor():
    num = int(input("숫자 "))
    d2 = num%2
    d3 = num%3
    d5 = num%5
    d8 = num%8
    tmp = [d2, d3, d5, d8]
    tmp
    
    factor=[]
    for n, d in zip([2, 3, 5, 8], tmp):
        if d == 0:
            factor.append(n)
    
    if len(factor) == 0:
        print(f'{num}은 2, 3, 5, 8의 배수가 아닙니다.')
    else:
        print(f'{num}은 {list((factor[i]) for i in range(len(factor)))}의 배수입니다.') # list >ragnge(leng(factor))

factor()
```
