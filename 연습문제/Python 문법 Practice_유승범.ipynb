## for문을 사용하여 50 이하의 4의 배수를 출력하시오.

## while문을 사용하여 1에서 100 사이의 홀수의 합을 구하시오. 
- output : 2500

```python
num = 1
output = 0

while num <= 100:
    if num%2 == 1:
        output += num
    num = num + 1

output
```

## for문을 사용하여 li7의 평균점수, 최저점수, 최고점수를 구하시오. 
- li7 = [80,82,82,82,89,85,94,95,99,100,47,70,80,84,92,92]
- output : 


```python
li7 = [80,82,82,82,89,85,94,95,99,100,47,70,80,84,92,92]
min1 = 1000
max1 = 0
sum1 = 0

for i in li7:
    sum1 += i
    if min1 > i:
        min1 = i
    if max1 < i:
        max1 = i

print('평균 :', sum1/len(li7))
print('최저점수 :', min1)
print('최고점수 :', max1)
```

## 파라미터 a,b중 더 큰 수를 리턴하는 함수(num)를 작성하시오. (만약 두 수가 같다면 0을 리턴) 



```python
def num(a, b):
    if a > b:
        return a
    if a < b:
        return b
    if a == b:
        return 0
```

```python
num(3,7)
```

```python
num(10,4)
```

```python
num(6,6)
```

## 함수를 사용하여 사용자에게 인풋을 받아서 반별로 자기소개 딕셔너리를 만드시오.
- 학생수, 이름, 나이, 성별
- num = 학생수 




```python


def roster (num):
    a = {}
    for i in range(num):
        name = input("이름")
        age = input("나이")
        sex = input("성별")

        a[name] = {}
        a[name]['나이 '] = age
        a[name]['성별 '] = sex
    print(a)
    
```

```python
roster(3)
```
