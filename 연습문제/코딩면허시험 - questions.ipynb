```python
# + 를 사용하면 기존의 리스트 a 값이 변하는 게 아니라 새로운 리스트가 반환된다.
# extend를 이용하면 주소 값이 변하지 않고 그대로 유지된다.

a = [1, 2, 3]
print(id(a))

# + 사용
a =  a + [4, 5]
print(a)
print(id(a))

a = [1, 2, 3]
print(id(a))

# extend 사용
a.extend([4, 5])
print(a)
print(id(a))
```

### Q11) 오류와 예외 처리

```python
result = 0

try:
    [1, 2, 3][3]   
    'a' + 1
    4 / 0
except TypeError:
    result += 1
except ZeroDivisionError:
    result += 2
except IndexError:  
    result += 3
finally:           
    result += 4

#print(result)
```

```python
def compress_string(s):
    _c = ""
    cnt = 0
    result = ""
    for c in s:
        if c != _c:
            _c = c
            if cnt: result += str(cnt)
            result += c
            cnt = 1
        else:
            cnt +=1
    if cnt: result += str(cnt)
    return result

compress_string("aaabbbccaadd")
```

### Q15) Duplicate Numbers

```python
# 숫자: 0123456789, 01234, 01234567890, 6789012345, 012322456789
def DupNum(num):
    lena = len(num)
    a = set(num)
    lenb = len(a)

    if lenb == 10 and lena == 10:
        print("True")
    else:
        print("False")

num = input()
#slicenum = num.slice(",")
DupNum(num)
```
